```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
single_cart_disp = pd.read_csv('single_agent_upright/travel_logs.csv', header = None)
single_cart_disp.columns = ['disp']

plt.plot(single_cart_disp)
plt.xlabel('Time')
plt.ylabel('Displacement From Center (px)')

print(single_cart_disp.mean(axis=0))
```

```python
mas_mdqn_cart1_disp = pd.read_csv('multi-agent-upright-multi-dqn-cart1/travel_logs.csv', header = None)
mas_mdqn_cart1_disp.columns = ['disp']

mas_mdqn_cart2_disp = pd.read_csv('multi-agent-upright-multi-dqn-cart2/travel_logs.csv', header = None)
mas_mdqn_cart2_disp.columns = ['disp']

plt.plot(mas_mdqn_cart1_disp, 'r')
plt.plot(mas_mdqn_cart2_disp, 'b')
plt.xlabel('Time')
plt.ylabel('Displacement From Center (px)')
print(mas_mdqn_cart1_disp.mean(axis=0))
print(mas_mdqn_cart2_disp.mean(axis=0))
```
